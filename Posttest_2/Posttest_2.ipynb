```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('movie_data.csv')

#1. Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.
desc_stats = df.describe().T
print("Statistik Deskriptif:")
print(desc_stats)

```

```python
# 2. Tampilkan 10 record pertama dari dataset dengan attribute bertipe angka.
numeric_df = df.select_dtypes(include=[np.number])
print("\n10 record pertama atribut numerik:")
print(numeric_df.head(10))
```

```python
# 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.
print("\nJumlah Record:", df.shape[0])
print("Jumlah attribute:", df.shape[1])
print("Nama attribute:", list(df.columns))
print("Jumlah attribute numerik:", len(numeric_df.columns))
```

```python
# 4. Tampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).
object_df = df.select_dtypes(include=["object"])
for col in object_df.columns:
    print(f"\nAttribute: {col}")
    print("Jumlah label unik:", object_df[col].nunique())
    print("Label unik:", object_df[col].unique())
```

```python
# 5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.
plt.figure(figsize=(8,5))
sns.histplot(df[numeric_df.columns[0]], kde=True, bins=20)
plt.title(f"Distribusi {numeric_df.columns[0]}")
plt.show()
```

```python
# 6. Tampilkan korelasi dataset menggunakan heatmap.
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Antar Atribut Numerik")
plt.show()
```

# Kesimpulan Analisis Data
1. Dataset terdiri dari 250 film dengan 8 atribut, di mana 2 atribut numerik('year' dan 'rating').
2. Statistik deskriptif menunjukkan:
   - Film terbanyak berasal dari tahun 2000-2020.
   - Rating film rata-rata cukup rendah, yaitu sekitar 3,58 dengan rentang dari 1.2 hingga 4.8. Setengah dari film(Q1 & Q3) memiliki rating antara 3.2 dan 4.2. Standar deviasi yang kecil (0.81).
3. Atribut kategorikal memiliki variasi label yang cukup tinggi, misalnya genre(99 jenis) dan sutradara (215 orang).
4. Distribusi rating cenderung normal dengan puncak di sekitar 3.5-4.0.
5. Korelasi antara 'year' dan 'rating' sangat rendah, artinya tahun rilis film tidak berpengaruh besar pada penilaian rating.
6. Dengan banyaknya variasi genre dan sutradara, dataset ini berpotensi digunakan untuk analisis lebih lanjut, misalnya:
    - Rekomendasi film berdasarkan genre.
    - Tren kualitas berdasarkan tahun.
    - Analisis pengaruh sutradara terhadap rating.
